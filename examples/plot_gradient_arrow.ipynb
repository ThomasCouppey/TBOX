# Plot gradient arrow

```python
import sys
sys.path.append("../")

from tbox.plot_utils import rainbowarrow
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np


fig,ax = plt.subplots()
ax.axis([0,5,0,4])
ax.set_aspect("equal")
N =200
vals = np.ones((N, 4))
vals[:, 0] *= 0
vals[:, 1] *= 0
vals[:, 2] *= np.linspace(0, 0, N)
vals[:, 3] *= np.linspace(1, .1, N)
cmap =  ListedColormap(vals)
rainbowarrow(ax, (3,3), (2,2.5), cmap="viridis", n=100,lw=3)
rainbowarrow(ax, (4,1), (1.5,1.5), cmap=((0,0,.8,0.1),(0,.6,0,.9)), n=300,lw=5,headw=None)
rainbowarrow(ax, (1,1), (1,2), cmap=cmap, n=N,lw=2, headw=60)

plt.show()
```
